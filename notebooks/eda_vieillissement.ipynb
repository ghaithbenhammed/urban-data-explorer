```python
import pandas as pd

df = pd.read_csv("../data/bronze/population_agee.csv",sep=";")

print(df.head())

print(df.info())

print(df.columns)

print("\nAGEMEN7 :")
print(
    df["AGEMEN7"]
    .value_counts()
    .sort_index()
)

print("\nTYPLR :")
print(
    df["TYPLR"]
    .value_counts()
    .sort_index()
)
print(df["CODGEO"].head(30))
print(df["CODGEO"].unique()[:30])


paris = df[
    df["CODGEO"].astype(str).str.startswith("751")
]

print(paris.head())

print(
    paris["CODGEO"]
    .unique()
)
```
