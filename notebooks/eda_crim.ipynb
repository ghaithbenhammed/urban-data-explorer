```python
import pandas as pd

# =========================
# 1. LOAD PARQUET
# =========================

print("Chargement données criminalité...")

df = pd.read_parquet(
    "../data/bronze/criminalite.parquet"
)

# =========================
# 2. APERÇU
# =========================

print(df.head())

# =========================
# 3. INFORMATIONS
# =========================

print(df.info())

# =========================
# 4. COLONNES
# =========================

print("Colonnes disponibles :")

for col in df.columns:
    print(col)

# =========================
# 5. TAILLE DATASET
# =========================

print("Shape dataset :")
print(df.shape)

# =========================
# 6. TYPES D'INFRACTIONS
# =========================

print("Types d'infractions :")

print(df["indicateur"].value_counts())

# =========================
# 7. ANNÉES DISPONIBLES
# =========================

print("Années disponibles :")

print(df["annee"].unique())

# =========================
# 8. FILTRE PARIS
# =========================

print("Filtre Paris...")

paris = df[
    df["CODGEO_2025"].astype(str).str.startswith("751")
]

print(paris.head())

print("Nombre lignes Paris :")
print(paris.shape)

# =========================
# 9. ARRONDISSEMENTS PARIS
# =========================

print("Arrondissements présents :")

print(
    sorted(
        paris["CODGEO_2025"].unique()
    )
)

# =========================
# 10. TAUX CRIMINALITÉ
# =========================

print("Taux criminalité moyen :")

print(
    paris.groupby("CODGEO_2025")[
        "taux_pour_mille"
    ].mean()
)

# =========================
# 11. INFRACTIONS LES PLUS FRÉQUENTES
# =========================

print("Infractions les plus fréquentes :")

print(
    paris.groupby("indicateur")[
        "nombre"
    ].sum().sort_values(ascending=False).head(10)
)
```
