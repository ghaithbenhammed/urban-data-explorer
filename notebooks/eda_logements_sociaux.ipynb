```python
import pandas as pd

print("Chargement RPLS...")

df = pd.read_csv(
    "../data/bronze/logements_sociaux.csv",
    sep=";",
    low_memory=False
)

# =========================
# APERÇU
# =========================

print(df.head())

# =========================
# INFOS
# =========================

print(df.info())

# =========================
# COLONNES
# =========================

print(df.columns)

# =========================
# FILTRE PARIS
# =========================

paris = df[
    df["Code Postal"]
    .astype(str)
    .str.startswith("75")
]

print(paris.head())

print(paris.shape)

# =========================
# ARRONDISSEMENTS
# =========================

paris["arrondissement"] = (
    paris["Code Postal"]
    .astype(str)
    .str[-2:]
)

print(
    paris["arrondissement"]
    .value_counts()
)
```
